```python
import os
from collections import Counter
import pickle
 

vowels=['अ','आ','इ','ई','उ','ऊ','ऋ','ए','ऐ','ओ','औ','अं','अः']

consonant=['क','ख','ग','घ','ङ','च','छ','ज','झ','ञ','ट','ठ','ड','ढ','ण','त','थ','द','ध',
           'न','प','फ','ब','भ','म','य','र','ल','व','श','ष','स','ह','क्ष','त्र','ज्ञ']

matra=['ा','ि','ी','ु','ू','ृ','ॄ','ॅ','े','ै','ॉ','ो','ौ','ं','ः','ँ','्र']

matra_mapping={'ा':'आ','ि':'इ','ी':'ई','ु':'उ','ू':'ऊ','े':'ए','ै':'ऐ','ो':'ओ','ौ':'औ','ृ':'ऋ',
               'ॄ':'ऋ','ॉ':'ओ','ॅ':'ए','ं':'अं','ँ':'अं','ः':'अः','्र':'र्','्':'अ'}

digits=['०','१','२','३','४','५','६','७','८','९','1','2','3','4','5','6','7','8','9','0']

punctuations=['।',':',';',',','.','\'','\\','/','-','‘','’','(',')','"','|']
```

```python
def divide(shabd):
    ans = []
    shabd+= 'प'
    i=0
    while(i<(len(shabd)-1)):
        if(shabd[i] in matra):
            ans.append(matra_mapping[shabd[i]])
        elif(shabd[i] in vowels):
            ans.append(shabd[i])
        elif(shabd[i] in consonant and shabd[i+1] in ['्','़']):
            ans.append(shabd[i]+ shabd[i+1])
        
        elif(shabd[i] in consonant and shabd[i+1] in matra):
            ans.append(shabd[i]+'्')
        elif(shabd[i] in consonant and shabd[i+1] in consonant+vowels):
            ans.append(shabd[i]+'्')
            ans.append('अ')
        
        i+=1
    return ans
```

```python
def ngram(word,ngram):
    #print(word)
    #shabds=divide(word)
    #print(shabds)
    temp=zip(*[shabds[i:] for i in range(0,ngram)])
    ans=[''.join(ngram) for ngram in temp]
    return ans
```

## Unigram

```python
unigram_char={}
f = 1
for i in os.listdir("data"):
    file = open('data/'+ str(i),encoding='utf-8', errors='ignore')
    for line in file:
        for j in punctuations+digits:
            line = line.replace(j," ")
        val = line.split()
        for i in val:
            shabds = divide(i)
            #print(shabds)
            if len(shabds)==0:
                continue
            words= ngram(shabds,1)
            #print(words)
            for k in words:
                if k in matra:
                    continue
                if k in ['',' ']:
                    continue
                if k in unigram_char.keys():
                    unigram_char[k] = unigram_char[k]+1
                else:
                    unigram_char[k]=1
    file.close()
    f+=1
    print(f)

```

```python
unigrams=sorted(unigram_char.items(), key=lambda x: x[1], reverse=True)[:100]
for i in unigrams:
    print(i[0])
    
unigram=[i[0] for i in unigrams]
    
with open('chars_unigram.pkl','wb') as f:
    pickle.dump(unigram, f)
    
with open("chars_unigrams.txt", 'w',encoding="utf-8") as output:
    for row in unigrams:
        output.write(str(row) +"\n")
```
