```python
import os
import pickle

vowels=['अ','आ','इ','ई','उ','ऊ','ऋ','ए','ऐ','ओ','औ','अं','अः']

consonent=['क','ख','ग','घ','ङ','च','छ','ज','झ','ञ','ट','ठ','ड','ढ','ण','त','थ','द','ध',
           'न','प','फ','ब','भ','म','य','र','ल','व','श','ष','स','ह','क्ष','त्र','ज्ञ']

matra=['ा','ि','ी','ु','ू','ृ','ॄ','ॅ','ॆ','े','ै','ॉ','ॊ','ो','ौ','्']

matra_mapping={'ा':'आ','ि':'इ','ी':'ई','ु':'उ','ू':'ऊ','े':'ए','ै':'ऐ','ो':'ओ','ौ':'औ','ृ':'ऋ',
               'ॄ':'ऋऋ','ॉ':'ओ','ॅ':'ए','ं':'अं','ँ':'अं','ः':'अः','ॆ':'ए','ॊ':'ओ','्र':' ','्':'अ'}

digits=['०','१','२','३','४','५','६','७','८','९','1','2','3','4','5','6','7','8','9','0']

punctuations=['।',':',';',',','.','\'','\\','/','-','‘','’','(',')','"','|']
```

```python
def ngrams(word,val):
    temp=zip(*[word[i:] for i in range(0,val)])
    ans=[' '.join(gram) for gram in temp]
    return ans
```

## Unigrams

```python
unigram_words={}
f = 1
for i in os.listdir("data"):
    file = open('data/'+ str(i),encoding='utf-8', errors='ignore')
    for line in file:
        for j in punctuations+digits:
            line = line.replace(j," ")
        val = line.split()
        words= ngrams(val,1)
        for k in words:
            if k in unigram_words.keys():
                unigram_words[k] = unigram_words[k]+1
            else:
                unigram_words[k]=1
    file.close()
    f+=1
    
unigram=sorted(unigram_words.items(), key=lambda x: x[1], reverse=True)[:100]
for i in unigram:
    print(i[0])
unigram=[i[0] for i in unigram]
    
with open('words_unigram.pkl','wb') as f:
    pickle.dump(unigram, f)
    
with open("words_unigram.txt", 'w',encoding="utf-8") as output:
    for row in unigram:
        output.write(str(row) + '\n')
```
